### Each state (node) of the search problem is represented with a tuple $((x, y), backpack, balls\_done)$ where
#### $(x, y)$ denotes the current location in the grid
#### $backpack$ is a binary string of the length of number of balls. $backpack[i] = 1 \Leftrightarrow$ ball number $i$ is in our backpack
#### $balls\_done$ is a binary string of the length of number of balls. $backpack = 1 \Leftrightarrow$ ball number $i$ is put in it's final location

```python
f = open("1.txt", "r")

lines = []
for line in f:
    l = line.rstrip("\n")
    lines.append(l)
    
print(lines)


n, m = map(int, lines[0].split())

start_x, start_y = map(int, list(lines[1].split()))

finish_x, finish_y = map(int, lines[2].split())

c = int(lines[3])

k = int(lines[4])

ball = []
for i in range(k):
    x1, y1, x2, y2 = map(int, lines[5 + i].split())
    ball.append((x1, y1, x2, y2))

    
grid = []
for i in range(n):
    row = ""
    for ch in lines[5 + k + i]:
        if ch != ' ':
            row += ch
    grid.append(row)
```

## BFS (Breadth First Search)

```python
import copy
import time



def print_path(cur):
    print(cur)
    if (cur == start_state):
          return
    print_path(par[cur])


start_state = ( (start_x, start_y), '0' * len(ball), '0' * len(ball) ) #loc, backpack, balls_done
goal_state =  ( (finish_x, finish_y), '0' * len(ball), '1' * len(ball) )

mark = dict()
mark[start_state] = 1
q = []
q.append(start_state)
dist = dict()
dist[start_state] = 0

par = dict()

while (len(q) > 0):
    cur = q.pop(0)
    
    if (cur == goal_state):
        break
    
    x, y = cur[0][0], cur[0][1]
    backpack = cur[1]
    balls_done = cur[2]
    
    
    #Put down a ball in no moves
    for i in range(len(backpack)):
        if (backpack[i] == '0'): #dont have it in our backpack
            continue
        if (ball[i][2] == x and ball[i][3] == y):
            backpack = backpack[:i] + "0" + backpack[i + 1:]
            balls_done = balls_done[:i] + "1" + balls_done[i + 1:]
                    
    
    #print(x, y, backpack, balls_done, dist[cur])
    
    
    #Pick up a ball
    for i in range(len(ball)):
        if (balls_done[i] == '1'): #Ball is already in final destination 
            continue
        if (backpack[i] == '1'): #Ball is already in backpack
            continue
            
        ballx, bally = ball[i][0], ball[i][1]
        if (ballx == x and bally == y):
            if (sum( map(int, list(backpack))) < c):
                new_backpack = copy.deepcopy(backpack)
                new_backpack = new_backpack[:i] + "1" + new_backpack[i + 1:] 
                
                new_state = ((x, y), new_backpack, balls_done)
                    
                if (not new_state in mark):
                    mark[new_state] = 1
                    q.append(new_state)
                    dist[new_state] = dist[cur] + 1
                    par[new_state] = cur
                
    
    #Go down
    if (x + 1 < n):
        if (grid[x + 1][y] != '*'):
            new_state = ((x + 1, y), backpack, balls_done)
            
            if (not new_state in mark):
                mark[new_state] = 1
                q.append(new_state)
                dist[new_state] = dist[cur] + 1
                par[new_state] = cur
    
    #Go up
    if (x - 1 >= 0):
        if (grid[x - 1][y] != '*'):
            new_state = ((x - 1, y), backpack, balls_done)
            
            if (not new_state in mark):
                mark[new_state] = 1
                q.append(new_state)
                dist[new_state] = dist[cur] + 1
                par[new_state] = cur
    
    #Go right
    if (y + 1 < m):
        if (grid[x][y + 1] != '*'):
            new_state = ((x, y + 1), backpack, balls_done)
                
            if (not new_state in mark):
                mark[new_state] = 1
                q.append(new_state)
                dist[new_state] = dist[cur] + 1
                par[new_state] = cur
    #Go left
    if (y - 1 >= 0):
        if (grid[x][y - 1] != '*'):
            new_state = ((x, y - 1), backpack, balls_done)
            
            if (not new_state in mark):
                mark[new_state] = 1
                q.append(new_state)
                dist[new_state] = dist[cur] + 1
                par[new_state] = cur
                

st = time.time()
print("The path len is {}".format(dist[goal_state]))
fn = time.time()
print("The path is:")
print_path(goal_state)
print("Number of states seen", len(dist))
print("Time consumed is %s seconds" %(fn - st))
```

### Advantages:
1. The solution will definitely found out by BFS If there is some solution.
2. BFS will never get trapped in a blind alley, which means unwanted nodes.
3. If there is more than one solution then it will find a solution with minimal steps.

### Disadvantages:
1. Memory Constraints As it stores all the nodes of the present level to go for the next level.
2. If a solution is far away then it consumes time.

## IDS (Iterative Deepening Search)

```python
import copy
import time


def print_path(cur):
    print(cur)
    if (cur == start_state):
          return
    print_path(par[cur])

seen = dict()

start_state = ( (start_x, start_y), '0' * len(ball), '0' * len(ball) ) #loc, backpack, balls_done
goal_state =  ( (finish_x, finish_y), '0' * len(ball), '1' * len(ball) )
dist = dict()
par = dict()

total_states = 0

def DLS(v, p, target, depth, limit):
    if (depth == limit):
        return
    global total_states
    total_states += 1
    par[v] = p
    if (v != p):
        dist[v] = dist[p] + 1
    seen[v] = 1
    if (v == target):
        return True
    
    
    x, y = v[0][0], v[0][1]
    backpack = v[1]
    balls_done = v[2]
    
    
    #Put down a ball in no moves
    for i in range(len(backpack)):
        if (backpack[i] == '0'): #dont have it in our backpack
            continue
        if (ball[i][2] == x and ball[i][3] == y):
            backpack = backpack[:i] + "0" + backpack[i + 1:]
            balls_done = balls_done[:i] + "1" + balls_done[i + 1:]
                    
    
    #nv = ((x, y), backpack, balls_done)
    #print(nv)
    #print(dist)
    #print(par)
    #print(seen)
    #print()
    
    
    #Pick up a ball
    for i in range(len(ball)):
        if (balls_done[i] == '1'): #Ball is already in final destination 
            continue
        if (backpack[i] == '1'): #Ball is already in backpack
            continue
            
        ballx, bally = ball[i][0], ball[i][1]
        if (ballx == x and bally == y):
            if (sum( map(int, list(backpack))) < c):
                new_backpack = copy.deepcopy(backpack)
                new_backpack = new_backpack[:i] + "1" + new_backpack[i + 1:] 
                
                new_state = ((x, y), new_backpack, balls_done)
                if (not new_state in seen):
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
                elif dist[new_state] > dist[v] + 1:
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
                
                
    
    #Go down
    if (x + 1 < n):
        if (grid[x + 1][y] != '*'):
            new_state = ((x + 1, y), backpack, balls_done)
            if (not new_state in seen):
                if (DLS(new_state, v, target, depth + 1, limit)):
                    del seen[v]
                    return True
            elif dist[new_state] > dist[v] + 1:
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
            
    
    #Go up
    if (x - 1 >= 0):
        if (grid[x - 1][y] != '*'):
            new_state = ((x - 1, y), backpack, balls_done)
            if (not new_state in seen):
                if (DLS(new_state, v, target, depth + 1, limit)):
                    del seen[v]
                    return True
            elif dist[new_state] > dist[v] + 1:
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
            
    
    #Go right
    if (y + 1 < m):
        if (grid[x][y + 1] != '*'):
            new_state = ((x, y + 1), backpack, balls_done)
            if (not new_state in seen):
                if (DLS(new_state, v, target, depth + 1, limit)):
                    del seen[v]
                    return True 
            elif dist[new_state] > dist[v] + 1:
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
            
    #Go left
    if (y - 1 >= 0):
        if (grid[x][y - 1] != '*'):
            new_state = ((x, y - 1), backpack, balls_done)
            if (not new_state in seen):
                if (DLS(new_state, v, target, depth + 1, limit)):
                    del seen[v]
                    return True   
            elif dist[new_state] > dist[v] + 1:
                    if (DLS(new_state, v, target, depth + 1, limit)):
                        #del seen[v]
                        return True
    
    #del seen[v]
    return False

def IDDFS(init, goal, max_depth):
    dist[init] = 0
    for i in range(max_depth + 1):
        print("Checking depth ", i)
        seen.clear()
        res = DLS(init, init, goal, 0, i) 
        if (res == True):
            return True
    return False

st = time.time()
res = IDDFS(start_state, goal_state, 100)
fn = time.time()
if (res):
    print("Path found with IDS")
    print("Len of path is ", dist[goal_state])
    print_path(goal_state)
    print("Time consumed is %s seconds" %(fn - st))
    print("Unique states seen", len(par))
    print("Total states seen", total_states)
    
else:
    print("Path not found with IDS")
```

# A* 

### The first heuristic we use is the manhattan distance: $manhattan(x_1, y_1, x_2, y_2) = |x_1 - x_2| + |y_1 - y_2|$

$h(node) = manhattan(node.x, node.y, x\_goal, y\_goal)$

### In each step the node with minimum $g(node) + h(node)$ is chosen where $g(node)$ is the cost of reaching this node from init state. 

#### In the implementation below, the parameter $g$ is not added to the state itself. It is added alonside each to state t $openset$ and $closedset$ like $(g, state)$

#### For example if we have a state $((2, 3), 010, 100)$ with cost $g = 11$ until now, the tuple $(11,((2, 3), 010, 100))$ is added to $openset$ and $closedset$

```python
from _heapq import heappush, heappop


import time

start_state = ( (start_x, start_y), '0' * len(ball), '0' * len(ball) ) #loc, backpack, balls_done
goal_state =  ( (finish_x, finish_y), '0' * len(ball), '1' * len(ball) )
par = dict()

def print_path(cur):
    print(cur)
    if (cur == start_state):
          return
    print_path(par[cur])

def manhattan(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)
    
def get_heuristic(node, goal):
    return manhattan(node[0][0], node[0][1], goal[0][0], goal[0][1])

    
is_in_heap = dict()
    
    
def aStar(start, goal):
    openset = []
    closedset = set([])
    explored = 0
    current = start
    
    print("Start is ", start)
    
    #openset.append((0, current))
    heappush(openset, (0 + get_heuristic(current, goal), current))
    is_in_heap[(0 + get_heuristic(current, goal), current)] = 1
    
    
    while openset:
        explored += 1
        current = heappop(openset)
        del is_in_heap[current]
            
        if current[1] == goal:    
            plen = 0
            node = current[1]
            while (node != start):
                node = par[node]
                plen += 1
            print("Path found with len", plen)
            print_path(goal_state)
            return explored
        
        closedset.add(current[1])
        
        x, y = current[1][0][0], current[1][0][1]
        backpack = current[1][1]
        balls_done = current[1][2]
        cur_gh = current[0]

        
        
        #Put down a ball in no moves
        for i in range(len(backpack)):
            if (backpack[i] == '0'): #dont have it in our backpack
                continue
            if (ball[i][2] == x and ball[i][3] == y):
                backpack = backpack[:i] + "0" + backpack[i + 1:]
                balls_done = balls_done[:i] + "1" + balls_done[i + 1:]
                
                closedset.add(((x, y), backpack, balls_done))
        
        for i in range(len(ball)):
            if (balls_done[i] == '1'): #Ball is already in final destination 
                continue
            if (backpack[i] == '1'): #Ball is already in backpack
                continue
            
            ballx, bally = ball[i][0], ball[i][1] #balls initial location
            if (ballx == x and bally == y):
                if (sum( map(int, list(backpack))) < c): #have enough capacity
                    new_backpack = copy.deepcopy(backpack)
                    new_backpack = new_backpack[:i] + "1" + new_backpack[i + 1:] 
                    
                    
                    new_state = ((x, y), new_backpack, balls_done)
                    
                    if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                        par[new_state] = current[1]
                        heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                        is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                        
        
        if (x + 1 < n):
            if (grid[x + 1][y] != '*'):
                
                new_state = ((x + 1, y), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                    
        if (x - 1 >= 0):
            if (grid[x - 1][y] != '*'):
                new_state = ((x - 1, y), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
       
        
        if (y + 1 < m):
            if (grid[x][y + 1] != '*'):
                new_state = ((x, y + 1), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
        
        if (y - 1 >= 0):
            if (grid[x][y - 1] != '*'):
                new_state = ((x, y - 1), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
        
st = time.time()
explored = aStar(start_state, goal_state)
fn = time.time()
print("Number of states explored is", explored)
print("Time consumed is %s" %(fn - st))
```

```python
from _heapq import heappush, heappop
from math import sqrt

import time

start_state = ( (start_x, start_y), '0' * len(ball), '0' * len(ball) ) #loc, backpack, balls_done
goal_state =  ( (finish_x, finish_y), '0' * len(ball), '1' * len(ball) )
par = dict()

def print_path(cur):
    print(cur)
    if (cur == start_state):
          return
    print_path(par[cur])

def manhattan(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)
    
def euclidean(x1, y1, x2, y2):
    return sqrt( (x1 - x2)**2 + (y1 - y2)**2 )
    
def get_heuristic(node, goal):
    return euclidean(node[0][0], node[0][1], goal[0][0], goal[0][1]) 
    
is_in_heap = dict()
    


    
def aStar(start, goal):
    openset = []
    closedset = set([])
    explored = 0
    current = start
    
    print("Start is ", start)
    
    #openset.append((0, current))
    heappush(openset, (0 + get_heuristic(current, goal), current))
    is_in_heap[(0 + get_heuristic(current, goal), current)] = 1
    
    
    while openset:
    
        current = heappop(openset)
        explored += 1
        del is_in_heap[current]
            
        if current[1] == goal:    
            plen = 0
            node = current[1]
            while (node != start):
                node = par[node]
                plen += 1
            print("Path found with len", plen)
            print_path(goal_state)
            return explored
        
        closedset.add(current[1])
        
        x, y = current[1][0][0], current[1][0][1]
        backpack = current[1][1]
        balls_done = current[1][2]
        cur_gh = current[0]

        
        
        #Put down a ball in no moves
        for i in range(len(backpack)):
            if (backpack[i] == '0'): #dont have it in our backpack
                continue
            if (ball[i][2] == x and ball[i][3] == y):
                backpack = backpack[:i] + "0" + backpack[i + 1:]
                balls_done = balls_done[:i] + "1" + balls_done[i + 1:]
                
                closedset.add(((x, y), backpack, balls_done))
        
        for i in range(len(ball)):
            if (balls_done[i] == '1'): #Ball is already in final destination 
                continue
            if (backpack[i] == '1'): #Ball is already in backpack
                continue
            
            ballx, bally = ball[i][0], ball[i][1] #balls initial location
            if (ballx == x and bally == y):
                if (sum( map(int, list(backpack))) < c): #have enough capacity
                    new_backpack = copy.deepcopy(backpack)
                    new_backpack = new_backpack[:i] + "1" + new_backpack[i + 1:] 
                    
                    
                    new_state = ((x, y), new_backpack, balls_done)
                    
                    if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                        par[new_state] = current[1]
                        heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                        is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                        
        
        if (x + 1 < n):
            if (grid[x + 1][y] != '*'):
                
                new_state = ((x + 1, y), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                    
        if (x - 1 >= 0):
            if (grid[x - 1][y] != '*'):
                new_state = ((x - 1, y), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
       
        
        if (y + 1 < m):
            if (grid[x][y + 1] != '*'):
                new_state = ((x, y + 1), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
        
        if (y - 1 >= 0):
            if (grid[x][y - 1] != '*'):
                new_state = ((x, y - 1), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
        
st = time.time()
explored = aStar(start_state, goal_state)
fn = time.time()
print("Number of states explored is", explored)
print("Time consumed is %s" %(fn - st))
```

# Weighted A*
### We modify the function above by adding a paremeter alpha and choosing the node in $openset$ which has the minimum $g(node) + \alpha\times h(node)$ where $h(node)$ is the manhattan distance of the node with goal state.

```python
from _heapq import heappush, heappop
from math import sqrt

import time

start_state = ( (start_x, start_y), '0' * len(ball), '0' * len(ball) ) #loc, backpack, balls_done
goal_state =  ( (finish_x, finish_y), '0' * len(ball), '1' * len(ball) )
par = dict()

def print_path(cur):
    print(cur)
    if (cur == start_state):
          return
    print_path(par[cur])

def manhattan(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)
    
def euclidean(x1, y1, x2, y2):
    return sqrt( (x1 - x2)**2 + (y1 - y2)**2 )
    
def get_heuristic(node, goal):
    global alpha
    return alpha * manhattan(node[0][0], node[0][1], goal[0][0], goal[0][1])
    
is_in_heap = dict()
    
def aStar(start, goal):
    
    openset = []
    closedset = set([])
    explored = 0
    current = start
    
    print("Start is ", start)
    
    #openset.append((0, current))
    heappush(openset, (0 + get_heuristic(current, goal), current))
    is_in_heap[(0 + get_heuristic(current, goal), current)] = 1
    
    
    while openset:
    
        current = heappop(openset)
        explored += 1
        del is_in_heap[current]
            
        if current[1] == goal:    
            plen = 0
            node = current[1]
            while (node != start):
                node = par[node]
                plen += 1
            print("Path found with len", plen)
            print_path(goal_state)
            return explored
        
        closedset.add(current[1])
        
        x, y = current[1][0][0], current[1][0][1]
        backpack = current[1][1]
        balls_done = current[1][2]
        cur_gh = current[0]

        
        
        #Put down a ball in no moves
        for i in range(len(backpack)):
            if (backpack[i] == '0'): #dont have it in our backpack
                continue
            if (ball[i][2] == x and ball[i][3] == y):
                backpack = backpack[:i] + "0" + backpack[i + 1:]
                balls_done = balls_done[:i] + "1" + balls_done[i + 1:]
                
                closedset.add(((x, y), backpack, balls_done))
        
        for i in range(len(ball)):
            if (balls_done[i] == '1'): #Ball is already in final destination 
                continue
            if (backpack[i] == '1'): #Ball is already in backpack
                continue
            
            ballx, bally = ball[i][0], ball[i][1] #balls initial location
            if (ballx == x and bally == y):
                if (sum( map(int, list(backpack))) < c): #have enough capacity
                    new_backpack = copy.deepcopy(backpack)
                    new_backpack = new_backpack[:i] + "1" + new_backpack[i + 1:] 
                    
                    
                    new_state = ((x, y), new_backpack, balls_done)
                    
                    if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                        par[new_state] = current[1]
                        heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                        is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                        
        
        if (x + 1 < n):
            if (grid[x + 1][y] != '*'):
                
                new_state = ((x + 1, y), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
                    
        if (x - 1 >= 0):
            if (grid[x - 1][y] != '*'):
                new_state = ((x - 1, y), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
       
        
        if (y + 1 < m):
            if (grid[x][y + 1] != '*'):
                new_state = ((x, y + 1), backpack, balls_done)
                
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
        
        if (y - 1 >= 0):
            if (grid[x][y - 1] != '*'):
                new_state = ((x, y - 1), backpack, balls_done)
                if (not new_state in closedset) and (not (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state) in is_in_heap):
                    par[new_state] = current[1]
                    heappush(openset, (cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state))
                    is_in_heap[(cur_gh + 1 + get_heuristic(new_state, goal) - get_heuristic(current[1], goal), new_state)] = 1
 
alpha = 2
st = time.time()
explored = aStar(start_state, goal_state)
fn = time.time()
print("Number of states explored is", explored)
print("Time consumed is %s" %(fn - st))
```
